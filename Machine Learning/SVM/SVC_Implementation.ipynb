```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Custom Synthetic Data Points
from sklearn.datasets import make_classification

X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,
                        n_clusters_per_class=1,n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
pd.DataFrame(X)[1]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
## Importing SVC Library 

from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

**Linear Kernel:** Used when data is linearly separable. The decision boundary is a straight line or a hyperplane.

```python
svc.fit(X_train,y_train)
```

```python
## Predictions

y_pred=svc.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

### Lets Try with some different Kernel!

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,
                        n_clusters_per_class=2,n_redundant=0)
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1],hue=y)
```

See, This Scatter plot isn't lineary separable so Linear Kernel will return some bad accuracy!

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10)
```

```python
rbf=SVC(kernel='rbf')
```

**Radial Basis Function (RBF) Kernel:** A popular choice for non-linear data, using a Gaussian function to measure similarity based on distance.

```python
rbf.fit(X_train,y_train)
```

```python
## Predictions
y_pred=svc.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

### What if we use Polynomial Kernel?

**Polynomial Kernel:** Captures non-linear relationships by raising the input features to a polynomial degree.

```python
poly=SVC(kernel='poly')
poly.fit(X_train,y_train)
## Predictions
y_pred=poly.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

## Trying out Sigmoid Kernel!

**Sigmoid Kernel:** Works like a neural network's activation function; rarely used, but can model non-linear boundaries.

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
## Predictions
y_pred=sigmoid.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

Conclusion : Out of all the kernels we used RBF gave best result !

### Lets do hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV

## Defining parameter range
param_grid={'C':[0.1,1,10,100,1000],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'kernel':['rbf','linear','polynomial']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,cv=5,verbose=3,refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
## Predictions
y_pred=grid.predict(X_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```
