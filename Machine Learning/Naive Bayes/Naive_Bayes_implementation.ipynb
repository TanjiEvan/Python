In this case we can notice some class imbalance as ham being 70% of the dataset. This scenarios provide less accuracy so we'll have to deal with class imbalance first. which will be discussed in next sections

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.Message,df.Category, test_size=0.2, random_state=5)
```

```python
X_train
```

One-hot encoding is a technique used to convert categorical variables into a binary matrix, where each unique category becomes a separate column, and the presence of that category is indicated by a 1 (while others are 0). This helps in machine learning models that require numerical input.

```python
from sklearn.feature_extraction.text import CountVectorizer
 
v=CountVectorizer()
X_train_count= v.fit_transform(X_train.values)
X_test_count=v.transform(X_test.values) 
```

This code uses the `CountVectorizer` from `sklearn` to convert text data (from `X_train` and `X_test`) into a matrix of token counts, which is a form of one-hot encoding for text.

Here's a breakdown:

- `CountVectorizer()` creates an instance of the vectorizer that will transform text data into a sparse matrix of token counts.
- `v.fit_transform(X_train.values)` learns the vocabulary from the training data and transforms the text into a sparse matrix of word counts for the training data (`X_train_count`).
- `v.transform(X_test.values)` applies the learned vocabulary to the test data and transforms it similarly into token counts for the test data (`X_test_count`).

In short, it's preparing the text data for input into machine learning models.

Imagine you have a collection of text data in `X_train` and `X_test`, like sentences or paragraphs. But machine learning models can't understand words directly—they need numbers. So, we use `CountVectorizer`, which acts like a word counter.

1. First, `v.fit_transform(X_train.values)` goes through all the text in `X_train`, finds all the unique words (vocabulary), and then counts how many times each word appears in each text. It stores this count as a matrix of numbers.
   
2. Then, `v.transform(X_test.values)` looks at the text in `X_test` but only counts the words it already learned from `X_train`. It creates another matrix of word counts for the test data.

This way, both your training and test text are turned into a format that the machine learning model can understand.

Let's say you have two texts in `X_train`:

- "I love pizza"
- "Pizza is great"

Using `CountVectorizer`, it will create a vocabulary of unique words: `["I", "love", "pizza", "is", "great"]`. Then it counts how many times each word appears in the sentences:

- "I love pizza" becomes `[1, 1, 1, 0, 0]` (1 "I", 1 "love", 1 "pizza", 0 "is", 0 "great")
- "Pizza is great" becomes `[0, 0, 1, 1, 1]` (0 "I", 0 "love", 1 "pizza", 1 "is", 1 "great")

When you use `v.transform(X_test)`, it applies the same counting process but only for the vocabulary it learned from `X_train`.

Now for text input data we use Multinomial Navie Nayes

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
from sklearn.metrics import classification_report

y_pred = model.predict(X_test_count)

report = classification_report(y_test, y_pred)
print(report)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
from matplotlib import pyplot as plt

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt="d", xticklabels=['Not Spam', 'Spam'], yticklabels=['Not Spam', 'Spam'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix Visualization')
plt.show()
```
