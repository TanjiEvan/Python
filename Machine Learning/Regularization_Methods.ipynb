```python
import pandas as pd

df = pd.read_csv("dataset.csv")
df.head()
```

```python
X = df.drop("target", axis=1)
Y = df["target"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, Y_train)

linear_score = linear_model.score(X_test, Y_test) ## Provides R^2 value
print("Linear Regression Score:", linear_score)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

y_pred = linear_model.predict(X_test)
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)
mse, r2
## The r2 values are same !!
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=1.0) ## Lambda=1
lasso_model.fit(X_train, Y_train)

lasso_score = lasso_model.score(X_test, Y_test)
print("Lasso Regression Score:", lasso_score)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, Y_train)

ridge_score = ridge_model.score(X_test, Y_test)
print("Ridge Regression Score:", ridge_score)
```

```python
print("Linear Regression Coefficients:", linear_model.coef_)
```

```python
print("Lasso Regression Coefficients:", lasso_model.coef_)
## 0 means that feature is getting null
```

```python
print("Ridge Regression Coefficients:", ridge_model.coef_)
```
