## **Calculation**

```python
q1={"Australia" : 80 , "Germany" : 90 , "USA" : 100 , "China" : 60}
q2={"Brazil" : 80 , "India" : 100 , "China" : 40,  "Germany" : 90 }
```

```python
sales_q1 = pd.Series(q1)
sales_q1
```

```python
sales_q2 = pd.Series(q2)
sales_q2
```

```python
sales_q1 + sales_q2
```

```python
sales_q1.add(sales_q2 , fill_value=0) ## To handle NaN
```

```python
sales_q2.add(sales_q1 , fill_value=0)
```

```python
sales_q1 * 10
```

```python
sales_q1 +  10
```

```python
sales_q2 +  10
```

```python
sales_q1 // sales_q2
```

```python
sales_q1 [ "China"] + sales_q2 ["China"]
```
