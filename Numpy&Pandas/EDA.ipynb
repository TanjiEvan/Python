## Working with Excel datasheet , ( movies_db )

```python
df_movies = pd.read_excel("movies_db.xlsx" , sheet_name="movies" )
df_movies.head()
```

```python
df_financials = pd.read_excel("movies_db.xlsx" , sheet_name="financials" )
df_financials.head()
```

```python
df_financials["currency"].unique()
```

## Normalization 

```python
def std_curr(x) :
    if x=="$$" or x=="Dollars" :
        return "USD"
    else:
        return x
    
df_financials = pd.read_excel("movies_db.xlsx" , sheet_name="financials" ,
                              converters={"currency": std_curr})

df_financials.head()
     
```

## Data Joining 

```python
df_merged = pd.merge(df_movies , df_financials, on="movie_id")
df_merged
```

## Exporting in Excel

```python
df_merged.to_excel("Movies_merged.xlsx" , sheet_name="merged_movies",index=False)
```

## Handling Missing Values

```python
df.head()
```

```python
new_df = df.fillna(0) ##Replaces the NaN values with (...)
new_df.head()
```

```python
df.fillna({
    "temperature" : df.temperature.mean(),
    "windspeed" :   df.windspeed.max(),
    "event" : "No event"
})
```

```python
df.head()
```

```python
df.ffill() ##Forward fill, NaN value gets repleced by immediate last value
```

```python
df.bfill() ##backward fill,NaN value gets repleced by immediate next value
```

```python
df.dropna() #drops whole row with any NaN
```

```python
df.dropna(how="all") ## if whole row has NaN,then drop
```

```python
df.dropna(thresh=3) ##minimum how many col will have values
```
