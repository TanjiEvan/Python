# Feature Selection Approaches from Research Paper

## Overview
This implementation aims to replicate and test the feature selection methods described in the paper **"Determining Threshold Value on Information Gain Feature Selection to Increase Speed and Prediction Accuracy of Random Forest."** The selected features will later be compared for their effectiveness in improving model accuracy and reducing computation time.

## Objectives
1. **Proposed Threshold Method**: 
   - Calculate Information Gain (IG) for each feature and use the **standard deviation of IG values** as the threshold to select features.
   - Purpose: Dynamically adapt the threshold to the dataset characteristics, as proposed in the paper.

2. **Fixed Threshold Method**: 
   - Select features with IG values greater than or equal to a fixed threshold (e.g., 0.05).
   - Purpose: Serve as a traditional baseline for comparison.

3. **Correlation-Based Feature Selection (CBFS)**:
   - Use Random Forest to compute feature importance and select features with importance scores above the mean.
   - Purpose: Model-based approach for feature selection, included for comparison as per the paper.

4. **FFT Transformation with Proposed Threshold**:
   - Transform the dataset to the frequency domain using Fast Fourier Transform (FFT), compute IG, and apply the **proposed threshold**.
   - Purpose: Test whether the frequency domain representation improves feature selection.

## Why These Approaches?
1. The **Proposed Threshold Method**:
   - Hypothesized to improve feature selection by dynamically adapting the threshold to dataset characteristics.
   - However, the paper notes that this method might **not always work effectively**, especially after data transformations like FFT, where accuracy could degrade.

2. The **Fixed Threshold Method** serves as a traditional baseline, providing a simpler but sometimes effective approach for comparison.

3. **CBFS** offers a model-based selection method that considers feature interactions and relationships, giving insights into its effectiveness compared to IG-based methods.

4. The **FFT Transformation** tests whether transforming data to the frequency domain improves the identification of relevant features or worsens it, as observed in some scenarios in the paper.

## **Next Steps**
1. Implement each approach to extract selected features.
2. Compare the selected features across approaches in terms of count and relevance.
3. Use the selected features to train models (later) and evaluate their accuracy and computational efficiency.


---

## **Step 1: Proposed Threshold Method**

**Why?**

-Use the **standard deviation** of IG values as the threshold to dynamically select features.

-This is the core method proposed in the paper.

```python
# Calculate the threshold as the standard deviation of IG values
proposed_threshold = ig_df['Information Gain'].std()

# Select features with IG >= threshold
selected_features_proposed = ig_df[ig_df['Information Gain'] >= proposed_threshold]['Feature'].values

print("\nFeatures Selected by Proposed Threshold Method:")
print(selected_features_proposed)
print(f"Number of Features: {len(selected_features_proposed)}")
```

## **Step 2: Fixed Threshold Method**

**Why?**

-Use a fixed threshold for IG values to select features.

-This serves as a traditional baseline for comparison.

```python
# Use the 50th percentile (median) as the fixed threshold
fixed_threshold = np.percentile(ig_df['Information Gain'], 50)  

# Select features with IG >= fixed threshold
selected_features_fixed = ig_df[ig_df['Information Gain'] >= fixed_threshold]['Feature'].values

print("\nFeatures Selected by Fixed Threshold Method (Dynamic Percentile):")
print(selected_features_fixed)
print(f"Number of Features: {len(selected_features_fixed)}")
```

## **Step 3: CBFS Method**

**Why?**

-Use Random Forest to evaluate feature importance and select features based on the mean importance.

-This model-based method helps capture complex feature interactions.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel

# Train Random Forest on the dataset
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
rf.fit(X, y)

# Apply CBFS: Select features with importance >= mean importance
cbfs = SelectFromModel(rf, threshold="mean", prefit=True)
selected_features_cbfs = X.columns[cbfs.get_support()]

print("\nFeatures Selected by CBFS Method:")
print(selected_features_cbfs)
print(f"Number of Features: {len(selected_features_cbfs)}")
```

## **Step 4: FFT Transformation with Proposed Threshold**

**Why?**

-Apply FFT to the dataset, compute IG values, and use the proposed threshold to select features.

-This tests if transforming the data into the frequency domain improves feature selection.

```python
from scipy.fft import fft

# Apply FFT to the dataset
X_fft = fft(X, axis=0).real  # Only keep the real part of the FFT transformation

# Compute IG values on FFT-transformed data
information_gain_fft = mutual_info_classif(X_fft, y, random_state=42)

# Calculate the threshold for FFT-transformed data
fft_threshold = np.std(information_gain_fft)

# Select features with IG >= FFT threshold
selected_features_fft = np.where(information_gain_fft >= fft_threshold)[0]
selected_features_fft_names = X.columns[selected_features_fft]

print("\nFeatures Selected by FFT with Proposed Threshold:")
print(selected_features_fft_names)
print(f"Number of Features: {len(selected_features_fft_names)}")
```

---

### **Benefits of Using FFT**
Fast Fourier Transform (FFT) converts data from the time or spatial domain to the **frequency domain**, allowing us to analyze patterns that are periodic or hidden in the original representation. By transforming features, we may:
- **Reveal Hidden Patterns:** Some relationships in the data might only become evident when analyzed in terms of frequencies.
- **Reduce Noise:** In some cases, FFT can emphasize dominant frequencies (important patterns) and suppress noise.
- **Explore Feature Relationships Differently:** Frequency domain representation may highlight correlations or dependencies that aren't obvious in the original dataset.
