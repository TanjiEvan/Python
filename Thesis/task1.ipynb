**Recalculating IG after dropping Top Feature Each Time :**

```python
from sklearn.feature_selection import mutual_info_classif
import pandas as pd

# Calculate initial Information Gain for each feature
information_gain = mutual_info_classif(X_train, y_train, discrete_features='auto', n_jobs=-1)
ig_df = pd.DataFrame({'Feature': X_train.columns, 'Information Gain': information_gain})
ig_df = ig_df.sort_values(by='Information Gain', ascending=False).reset_index(drop=True)
print("Initial Information Gain for each feature:")
print(ig_df.to_string(index=False))

# Create a copy of X_train to avoid modifying the original
X_temp = X_train.copy()
iteration = 1

# Iteratively drop the top feature and calculate Information Gain
while X_temp.shape[1] > 1:  # Stop when only one feature remains
    # Identify the feature with the highest IG from the previous iteration
    top_feature = ig_df.iloc[0]['Feature']
    top_ig = ig_df.iloc[0]['Information Gain']
    
    # Display the dropped feature and current iteration
    print(f"\nIteration {iteration}: Dropping feature '{top_feature}' with Information Gain: {top_ig:.6f}")
    
    # Drop the feature with the highest IG from the dataset
    X_temp = X_temp.drop(columns=top_feature)
    
    # Recalculate Information Gain after dropping the top feature
    information_gain = mutual_info_classif(X_temp, y_train, discrete_features='auto', n_jobs=-1)
    ig_df = pd.DataFrame({'Feature': X_temp.columns, 'Information Gain': information_gain})
    ig_df = ig_df.sort_values(by='Information Gain', ascending=False).reset_index(drop=True)
    
    # Display updated Information Gain after recalculation
    print("Information Gain after dropping the feature:")
    print(ig_df.to_string(index=False))
    
    # Increment the iteration counter
    iteration += 1
```

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
import graphviz


# Set up RandomForest with a single estimator to simulate a Decision Tree and enable parallel processing
treeclassifier = RandomForestClassifier(n_estimators=1, n_jobs=-1, random_state=42)
treeclassifier.fit(X_train, y_train)

# Extract the single decision tree from the RandomForest
estimator = treeclassifier.estimators_[0]

# Export the tree to DOT format
dot_data = export_graphviz(
    estimator,
    out_file=None,
    feature_names=X_train.columns,
    class_names=[str(cls) for cls in treeclassifier.classes_],
    filled=True,
    rounded=True,
    special_characters=True
)

# Render the tree using Graphviz
graph = graphviz.Source(dot_data)
graph.render("full_decision_tree", format="pdf")  # Save as high-resolution PDF
graph.view()  # Opens the PDF for detailed visualization
```
