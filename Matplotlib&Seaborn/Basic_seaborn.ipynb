multiple: How to layer the bars, can be 'layer', 'dodge', 'stack', or 'fill'.

data: Dataset for plotting. If x and y are specified, data should be a DataFrame.
x: Variable for x-axis (required for horizontal histograms).
y: Variable for y-axis (required for vertical histograms).
bins: Number of bins for the histogram.
binwidth: Width of each bin.
binrange: Lower and upper range of the bins.
discrete: If True, set the bins to integers.
cumulative: If True, create a cumulative histogram.
common_bins: If True, use common bins for multiple histograms.
common_norm: If True, use common normalization for multiple histograms.
multiple: How to layer the bars, can be "layer", "dodge", "stack", or "fill".
element: Choose between "bars", "step", and "poly".
fill: If True, fill the bars.
stat: Can be "count", "frequency", "density", or "probability".
hue: Grouping variable for splitting the data by color.
hue_order: Order to plot the hue levels.
palette: Colors to use for the different levels of the hue variable.
hue_norm: Normalization in data units for scaling the colormap applied to the hue variable.
weights: Weights for each observation.
multiple: How to layer the bars, can be 'layer', 'dodge', 'stack', or 'fill'.
kde: If True, compute a kernel density estimate to smooth the histogram.
log_scale: If True, set the scale of the axis to logarithmic.
legend: If True, add a legend.
ax: Matplotlib Axes object to draw the plot onto.

# Countplot

```python
df["division"].value_counts()
```

```python
plt.figure(figsize=(10,3))
sns.countplot(data=df,x="division")
```

```python
plt.figure(figsize=(10,5))

sns.countplot(
    data=df,
    x="division",
    hue="level of education",  # Splits the bars by level of education
    palette="Set2",  # Sets the color palette
    saturation=0.75,  # Adjusts the saturation level of the bars
    dodge=True,  # Separates the bars by hue
    edgecolor="black",  # Adds black borders to the bars
    linewidth=1.5,  # Sets the width of the lines around the bars
    alpha=0.8  # Sets the transparency level of the bars
    
)
```

# Barplot

```python
plt.figure(figsize=(10,3))

sns.barplot(data=df,x="division",y="salary")
```

```python
plt.figure(figsize=(10,3))
sns.barplot(
    data=df,
    x="division",
    y="salary",
    hue="level of education",  # Splits the bars by level of education
    palette="Set2",  # Sets the color palette
    #ci="sd",  # Uses standard deviation for error bars
    saturation=0.75,  # Adjusts the saturation level of the bars
    errcolor="black",  # Sets the color of the error bars
    errwidth=1,  # Sets the thickness of the error bars
    capsize=0.1,  # Sets the width of the caps on the error bars
    dodge=True  # Separates the bars by hue
)
```

confidence interval/Error Bar/Standard deviation measures the amount of variation or dispersion in a set of values.
