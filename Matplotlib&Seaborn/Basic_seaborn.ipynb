```python
import numpy as np 
import pandas as pd  
import matplotlib.pyplot as plt 
import seaborn as sns 
```

# Scatter Plot  

-> Used to decide Co-relation

```python
df=pd.read_csv("dm_office_sales.csv")
df.head()
```

```python
sns.scatterplot(df,x="sales",y="salary");
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(df,x="sales",y="salary");
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(df,x="sales",y="salary", hue="division"); ## Categorical
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(df,x="sales",y="salary", hue="salary"); ##Numerical
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(df,x="sales",y="salary", hue="division",
                s=200);
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(df,x="sales",y="salary", hue="division",
                size="work experience");
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(df,x="sales",y="salary", hue="division",
                size="work experience",
                alpha=0.4);
```

If there is too many datas in one point we use Alpha ^^^

```python
plt.figure(figsize=(12,8))

sns.scatterplot(df,x="sales",y="salary", hue="division",
                size="work experience",
                style="level of education");
```
