```python
import numpy as np 
import pandas as pd  
import matplotlib.pyplot as plt 
```

```python
x=np.arange(11)
y= x**2
print (x)
print ()
print (y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.title("This is my Chart")
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(x,y,"r")
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.title("This is my Chart")
plt.subplot(3,1,3)
plt.plot(y,x , "b")
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.title("This is my Chart")
plt.show()
```

```python
plt.subplot(1,3,1)
plt.plot(x,y,"r")
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.title("This is my Chart")
plt.subplot(1,3,3)
plt.plot(y,x , "b")
plt.xlabel("This is X axis")
plt.ylabel("This is Y axis")
plt.title("This is my Chart")
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,"r")
axes.set_xlabel("This is X axis")
axes.set_ylabel("This is Y axis")
axes.set_title("This is my Chart")
```
