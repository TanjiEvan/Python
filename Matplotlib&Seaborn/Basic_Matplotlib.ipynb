```python
x = np.arange(11)
y = x**2

fig, axes = plt.subplots(1, 2, figsize=(10, 4))  # Create 1 row and 2 columns of subplots

# Plot x vs y on the first subplot (axes[0])
axes[0].plot(x, y, "r")
axes[0].set_xlabel("X Axis")
axes[0].set_ylabel("Y Axis")
axes[0].set_title("x vs y")

# Plot y vs x on the second subplot (axes[1])
axes[1].plot(y, x, "b")
axes[1].set_xlabel("Y Axis")
axes[1].set_ylabel("X Axis")
axes[1].set_title("y vs x")

# Display the plot
plt.tight_layout()  # Adjust the layout
plt.show()
```

```python
fig.savefig("chart.png") ##saves the chart
```

```python

fig, axes = plt.subplots(1, 3, figsize = (12,4))

axes[0].plot(x, y*5, x, y**3)
axes[0].set_title("this is my default chart")



axes[1].plot(y, y*10, y, y**2)
axes[1].axis('tight') ##adjusts for other graphs
axes[1].set_title("this is my tight axes chart")


axes[2].plot(x, x**5, x, x**3)
axes[2].set_xlim([0,9])
axes[2].set_ylim([0, 100])
axes[2].set_title("this is custom axes range")
```

```python
fig2 = plt.figure()

pp=fig2.add_axes([0,0,1,1]) ##Full 640x480 size canvas 

pp.plot(x,x**2 ,"r", label = 'X Squared Line')
pp.plot(x,x**2,"o",label = 'Points' )
pp.legend(loc=0)
```

"." : point marker

"," : pixel marker

"o" : circle marker

"v" : triangle_down marker

"^" : triangle_up marker

"<" : triangle_left marker

">" : triangle_right marker

"1" : tri_down marker

"2" : tri_up marker

"3" : tri_left marker

"4" : tri_right marker

"s" : square marker

"p" : pentagon marker

"*" : star marker

"h" : hexagon1 marker

"H" : hexagon2 marker

"+" : plus marker

"x" : x marker

"D" : diamond marker

"d" : thin_diamond marker

"|" : vline marker

"_" : hline marker

'-'

solid line style

'--'

dashed line style

'-.'

dash-dot line style

':'

dotted line style

## Working with Dataset

```python
rain=pd.read_csv("labels.csv")
rain
```

```python
fig,axes=plt.subplots (2,1,figsize=(12,8))

axes[0].plot(rain["Label"],rain["July 2007"],"r")

axes[0].set_title("July 2007")

# Plot y vs x on the second subplot (axes[1])
axes[1].plot(rain["Label"], rain["July 2020"])

axes[1].set_title("July 2020")

plt.show()
```

```python
fig=plt.figure()

rf=fig.add_axes([0,0,1,1])

rf.plot(rain["Label"],rain["July 2007"],"r")
rf.plot(rain["Label"], rain["July 2020"])
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a figure and axis
fig = plt.figure()
rf = fig.add_axes([0, 0, 1, 1])

# Plot the data
rf.plot(rain["Label"], rain["July 2007"], label="July 2007", color="red")
rf.plot(rain["Label"], rain["July 2020"], label="July 2020", color="blue")

# Add labels and title
rf.set_xlabel("Label")
rf.set_ylabel("Rainfall")
rf.set_title("Comparison of Rainfall in July 2007 and July 2020")

# Add legend
rf.legend()

# Show the plot
plt.show()
```

```python
fig,axes=plt.subplots (2,1,figsize=(12,8))

axes[0].bar(rain["Label"],rain["July 2007"],color="red")

axes[0].set_title("July 2007")

# Plot y vs x on the second subplot (axes[1])
axes[1].scatter(rain["Label"], rain["July 2020"])

axes[1].set_title("July 2020")

plt.show()
```
