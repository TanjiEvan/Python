```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
sns.jointplot(x = 'math score', y= 'reading score', data =df);
```

```python
sns.jointplot(x = 'math score', y= 'reading score', kind="hex" , data =df);
```

```python
sns.jointplot(x = 'math score', y= 'reading score', kind="hist" , data =df);
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, corner= True);
```

```python
sns.pairplot(df, hue= 'gender', palette= 'magma');
```

```python
sns.catplot(x = 'gender', y = 'math score', data = df, kind= 'box');
```

```python
sns.catplot(x = 'gender', y = 'math score', data = df, col= 'lunch', kind= 'box');
```

```python
sns.catplot( y = 'math score', data = df, col= 'lunch', kind= 'box');
```

```python
sns.catplot(x = 'gender', y = 'math score', hue="parental level of education", data = df, kind= 'bar');
```

```python
sns.catplot(x = 'gender', y = 'math score', data = df, kind= 'violin');
```

```python
sns.FacetGrid(data=df, col='gender');
```

```python
sns.FacetGrid(data=df, col='lunch');
```

```python
sns.FacetGrid(data=df, col='gender', row = 'lunch');
```

```python
g = sns.FacetGrid(data=df, col='gender', row = 'lunch')
g = g.map(plt.scatter, "math score", "reading score")
g.add_legend();
```

```python
g = sns.FacetGrid(data=df, col='gender', row = 'lunch')
g = g.map(plt.bar, "math score", "reading score")
g.add_legend();
```

```python
g = sns.FacetGrid(data=df, col='gender', row = 'lunch')
g = g.map(plt.scatter, "math score", "reading score")
g.add_legend();

plt.subplots_adjust(hspace= 0.4, wspace=0.5);
```
