### 3. Handling Null Values with Mean ,Mode , Median

```python
mean_ps=ps.fillna({
    "Rating" : ps.Rating.mean().round(2),

})
mean_ps
```

```python
median_ps=ps.fillna({
    "Rating" : ps.Rating.median().round(2),

})
median_ps
```

```python
mode_ps = ps.fillna({
    "Rating": ps["Rating"].mode()[0]
})
mode_ps
```

### 10. Percentage of free apps in our data 

```python
dropped_ps.head()
```

```python
count_price_zero = (dropped_ps["Price"] == "0").sum()
print("Number of entries with price=0:", count_price_zero)
```

```python
# Calculate the total number of entries
total_entries = len(dropped_ps["Price"])

# Calculate the percentage of free apps
percentage_free_apps = (count_price_zero / total_entries) * 100

percentage_free_apps
```
